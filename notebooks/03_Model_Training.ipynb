```python
# ============================================
# CELL 1: Imports
# ============================================
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import joblib
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported!")
```

```python
# ============================================
# CELL 2: Load Processed Data
# ============================================
X_train_int = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\X_train_int_smote.npy')
y_train_int = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\y_train_int_smote.npy')
X_test_int  = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\X_test_int.npy')
y_test_int  = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\y_test_int.npy')

X_train_phi = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\X_train_phi.npy')
y_train_phi = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\y_train_phi.npy')
X_test_phi  = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\X_test_phi.npy')
y_test_phi  = np.load(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\y_test_phi.npy')

print("✅ All processed data loaded!")
print(f"\nIntrusion — Train: {X_train_int.shape}, Test: {X_test_int.shape}")
print(f"Phishing  — Train: {X_train_phi.shape}, Test: {X_test_phi.shape}")
```

```python
# ============================================
# CELL 3: Logistic Regression (Baseline)
# ============================================
print("Training Logistic Regression...")
print("⚠️ May take 3-5 minutes...")

lr_int = LogisticRegression(random_state=42, max_iter=1000)
lr_int.fit(X_train_int, y_train_int)
lr_int_score = accuracy_score(y_test_int, lr_int.predict(X_test_int))

lr_phi = LogisticRegression(random_state=42, max_iter=1000)
lr_phi.fit(X_train_phi, y_train_phi)
lr_phi_score = accuracy_score(y_test_phi, lr_phi.predict(X_test_phi))

print(f"\n✅ Logistic Regression Done!")
print(f"Intrusion Accuracy: {lr_int_score:.4f} ({lr_int_score*100:.2f}%)")
print(f"Phishing  Accuracy: {lr_phi_score:.4f} ({lr_phi_score*100:.2f}%)")

# Save models
joblib.dump(lr_int, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\lr_intrusion.pkl')
joblib.dump(lr_phi, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\lr_phishing.pkl')
print("✅ Models saved!")
```

```python
# ============================================
# CELL 4: Decision Tree (Baseline)
# ============================================
print("Training Decision Tree...")

dt_int = DecisionTreeClassifier(random_state=42)
dt_int.fit(X_train_int, y_train_int)
dt_int_score = accuracy_score(y_test_int, dt_int.predict(X_test_int))

dt_phi = DecisionTreeClassifier(random_state=42)
dt_phi.fit(X_train_phi, y_train_phi)
dt_phi_score = accuracy_score(y_test_phi, dt_phi.predict(X_test_phi))

print(f"\n✅ Decision Tree Done!")
print(f"Intrusion Accuracy: {dt_int_score:.4f} ({dt_int_score*100:.2f}%)")
print(f"Phishing  Accuracy: {dt_phi_score:.4f} ({dt_phi_score*100:.2f}%)")

# Save models
joblib.dump(dt_int, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\dt_intrusion.pkl')
joblib.dump(dt_phi, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\dt_phishing.pkl')
print("✅ Models saved!")
```

```python
# ============================================
# CELL 5: Random Forest
# ============================================
print("Training Random Forest...")
print("⚠️ This will take 5-10 minutes...")

rf_int = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_int.fit(X_train_int, y_train_int)
rf_int_score = accuracy_score(y_test_int, rf_int.predict(X_test_int))

rf_phi = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_phi.fit(X_train_phi, y_train_phi)
rf_phi_score = accuracy_score(y_test_phi, rf_phi.predict(X_test_phi))

print(f"\n✅ Random Forest Done!")
print(f"Intrusion Accuracy: {rf_int_score:.4f} ({rf_int_score*100:.2f}%)")
print(f"Phishing  Accuracy: {rf_phi_score:.4f} ({rf_phi_score*100:.2f}%)")

# Save models
joblib.dump(rf_int, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\rf_intrusion.pkl')
joblib.dump(rf_phi, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\rf_phishing.pkl')
print("✅ Models saved!")
```

```python
# ============================================
# CELL 6: XGBoost
# ============================================
print("Training XGBoost...")
print("⚠️ This will take 5-10 minutes...")

xgb_int = XGBClassifier(n_estimators=100, random_state=42, 
                         n_jobs=-1, eval_metric='mlogloss')
xgb_int.fit(X_train_int, y_train_int)
xgb_int_score = accuracy_score(y_test_int, xgb_int.predict(X_test_int))

xgb_phi = XGBClassifier(n_estimators=100, random_state=42,
                         n_jobs=-1, eval_metric='logloss')
xgb_phi.fit(X_train_phi, y_train_phi)
xgb_phi_score = accuracy_score(y_test_phi, xgb_phi.predict(X_test_phi))

print(f"\n✅ XGBoost Done!")
print(f"Intrusion Accuracy: {xgb_int_score:.4f} ({xgb_int_score*100:.2f}%)")
print(f"Phishing  Accuracy: {xgb_phi_score:.4f} ({xgb_phi_score*100:.2f}%)")

# Save models
joblib.dump(xgb_int, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\xgb_intrusion.pkl')
joblib.dump(xgb_phi, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\xgb_phishing.pkl')
print("✅ Models saved!")
```

```python
# ============================================
# CELL 7: RF + XGBoost Ensemble (Novel Model)
# ============================================
print("Training RF + XGBoost Ensemble...")
print("⚠️ This will take 10-15 minutes...")

# Intrusion Ensemble
ensemble_int = VotingClassifier(
    estimators=[
        ('rf', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)),
        ('xgb', XGBClassifier(n_estimators=100, random_state=42, n_jobs=-1, eval_metric='mlogloss'))
    ],
    voting='soft'
)
ensemble_int.fit(X_train_int, y_train_int)
ensemble_int_score = accuracy_score(y_test_int, ensemble_int.predict(X_test_int))

# Phishing Ensemble
ensemble_phi = VotingClassifier(
    estimators=[
        ('rf', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)),
        ('xgb', XGBClassifier(n_estimators=100, random_state=42, n_jobs=-1, eval_metric='logloss'))
    ],
    voting='soft'
)
ensemble_phi.fit(X_train_phi, y_train_phi)
ensemble_phi_score = accuracy_score(y_test_phi, ensemble_phi.predict(X_test_phi))

print(f"\n✅ Ensemble Done!")
print(f"Intrusion Accuracy: {ensemble_int_score:.4f} ({ensemble_int_score*100:.2f}%)")
print(f"Phishing  Accuracy: {ensemble_phi_score:.4f} ({ensemble_phi_score*100:.2f}%)")

# Save models
joblib.dump(ensemble_int, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\ensemble_intrusion.pkl')
joblib.dump(ensemble_phi, r'C:\Users\Kshaunish\cyber-xai-ml-project\models\saved_models\ensemble_phishing.pkl')
print("✅ Ensemble models saved!")
```

```python
# ============================================
# CELL 8: Save Results Summary
# ============================================
results = {
    'Model': ['Logistic Regression', 'Decision Tree', 
              'Random Forest', 'XGBoost', 'RF+XGBoost Ensemble'],
    'Intrusion_Accuracy': [lr_int_score, dt_int_score, 
                           rf_int_score, xgb_int_score, ensemble_int_score],
    'Phishing_Accuracy': [lr_phi_score, dt_phi_score,
                          rf_phi_score, xgb_phi_score, ensemble_phi_score]
}

results_df = pd.DataFrame(results)
results_df['Intrusion_Accuracy'] = results_df['Intrusion_Accuracy'].apply(lambda x: f"{x*100:.2f}%")
results_df['Phishing_Accuracy'] = results_df['Phishing_Accuracy'].apply(lambda x: f"{x*100:.2f}%")

print("📊 FINAL RESULTS TABLE")
print("=" * 55)
print(results_df.to_string(index=False))

results_df.to_csv(r'C:\Users\Kshaunish\cyber-xai-ml-project\data\processed\model_results.csv', index=False)
print("\n✅ Results saved to model_results.csv!")
```
